## N=2 Images

Load Images from Local Placement

```python
# open images as grayscale and stack on top of each others:
A = cv2.imread('images/movie_flicker1.png', 0)
B = cv2.imread('images/movie_flicker2.png', 0)

imgs = np.stack((A,B))
imgs.shape
```

```python
# compute original CDFs:
CDFs = []
for x in range(len(imgs)):
    CDFs.append(cumHist(imgs[x]))
CDFs = np.array(CDFs)

# Apply midway equalization:
res_imgs, res_CDFs = general_midway_equalization(imgs)

"""
Plots
"""
# Original images:
plt.figure(figsize=(15, 4))
for i in range(0, len(imgs)):
    plt.subplot(1, len(imgs), i+1)
    plt.imshow(imgs[i], cmap=cm.Greys_r, vmin=0, vmax=255)
    plt.axis('off')
    plt.colorbar(fraction=0.026, pad=0.04)
    plt.grid(False)
    plt.title('Original {}'.format(i+1, fontsize=14))

# Resulting images:
plt.figure(figsize=(15, 4))
for i in range(0, len(res_imgs)):
    plt.subplot(1, len(res_imgs), i+1)
    plt.imshow(res_imgs[i], cmap=cm.Greys_r, vmin=0, vmax=255)
    plt.axis('off')
    plt.colorbar(fraction=0.026, pad=0.04)
    plt.grid(False)
    plt.title('Resulting {}'.format(i+1, fontsize=14))

# CDFs:
cmap1 = plt.cm.RdBu(np.linspace(0,1,len(CDFs)))
cmap2 = plt.cm.autumn(np.linspace(0,1,len(res_CDFs)))
fig = plt.figure(figsize=(10,5))
for n in range(len(CDFs)):
    plt.plot(CDFs[n]*100, '-', color=cmap1[n], label='CDF original {}'.format(n+1))
    plt.plot(res_CDFs[n]*100, '--', color=cmap2[n], label='CDF resulting {}'.format(n+1))
plt.title("Gray Channel", fontsize=15);
plt.ylabel("Cumulative Percentage [%]", fontsize=14)
plt.xlabel("Pixel Value", fontsize=14)
plt.legend()
plt.show()
```
